```python
#Background Subtraction
import cv2

video_path = r"C:\Users\DELL\OneDrive\Desktop\Afraa\Images\traffic.mp4"
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

width, height = 640, 360
bg_subtractor = cv2.createBackgroundSubtractorMOG2()

while True:
    ret, frame = cap.read()
    
    if not ret:
        break

    frame_resized = cv2.resize(frame, (width, height))
    fg_mask = bg_subtractor.apply(frame_resized)
    fg_mask = cv2.GaussianBlur(fg_mask, (5, 5), 0)

    cv2.imshow("Resized Original Video", frame_resized)
    cv2.imshow("Resized Foreground Mask", fg_mask)

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
#Averaging
import cv2
import numpy as np

video_path = r"C:\Users\DELL\OneDrive\Desktop\Afraa\Images\traffic.mp4"
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

ret, frame = cap.read()
if not ret:
    print("Error: Could not read the first frame.")
    cap.release()
    exit()

width, height = 640, 360
frame_resized = cv2.resize(frame, (width, height))
avg_background = np.float32(frame_resized)
alpha = 0.02  

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_resized = cv2.resize(frame, (width, height))
    cv2.accumulateWeighted(frame_resized, avg_background, alpha)
    background = cv2.convertScaleAbs(avg_background)  
    foreground_mask = cv2.absdiff(frame_resized, background)
    gray_foreground = cv2.cvtColor(foreground_mask, cv2.COLOR_BGR2GRAY)
    _, binary_mask = cv2.threshold(gray_foreground, 50, 255, cv2.THRESH_BINARY)

    cv2.imshow("Original Frame", frame_resized)
    cv2.imshow("Estimated Background", background)
    cv2.imshow("Foreground Mask", binary_mask)

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python

#Delete scene change
import cv2
import numpy as np

video_path = r"C:\Users\DELL\OneDrive\Desktop\Afraa\Images\traffic.mp4"  
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

ret, frame = cap.read()
if not ret:
    print("Error: Could not read the first frame.")
    cap.release()
    exit()

width, height = 640, 360
frame_resized = cv2.resize(frame, (width, height))
avg_background = np.float32(frame_resized)
alpha = 0.02  
scene_change_threshold = 3000000  

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_resized = cv2.resize(frame, (width, height))
    cv2.accumulateWeighted(frame_resized, avg_background, alpha)
    background = cv2.convertScaleAbs(avg_background)  
    foreground_mask = cv2.absdiff(frame_resized, background)
    gray_foreground = cv2.cvtColor(foreground_mask, cv2.COLOR_BGR2GRAY)
    scene_change_score = np.sum(gray_foreground)
    scene_change_detected = scene_change_score > scene_change_threshold
    status_text = "Scene Change: YES" if scene_change_detected else "Scene Change: NO"
    cv2.putText(frame_resized, status_text, (20, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255) if scene_change_detected else (0, 255, 0), 2)

    cv2.imshow("Original Frame", frame_resized)
    cv2.imshow("Estimated Background", background)
    cv2.imshow("Foreground Mask", gray_foreground)

    if scene_change_detected:
        print("Scene Change Detected!")

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
