```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

img=cv2.imread("C:/Users/DELL/OneDrive/Desktop/lena.png")
print(img.shape)
img
```

```python
#Using Libraries
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

fig=plt.figure(figsize=(8,4))
img1,img2,img3=fig.add_subplot(131),fig.add_subplot(132),fig.add_subplot(133)

#RGB Image
img2.imshow(image_rgb)	
img2.set_title('RGB Image')

#BGR Image
img1.imshow(img)
img1.set_title('BGR Iamge')

#Grayscale Image
img3.imshow(image_gray,	cmap='gray')
img3.set_title('Grayscale Image')
print(image_gray.shape)


plt.show()
```

```python
#Mathematical way using	numpy
r,g,b=img[:,:,0],img[:,:,1],img[:,:,2]
gamma=1.04
r_const,g_const,b_const=0.2126,0.7156,0.0722
grayscale_image=r_const*r**gamma+g_const*g**gamma+b_const*b**gamma
image = np.stack([b, g, r], axis=-1)

fig=plt.figure()
image1,image2,image3=fig.add_subplot(131),fig.add_subplot(132),fig.add_subplot(133)

#BGR Image
image1.imshow(img)
image1.set_title('BGR Image')

#RGB Image
image2.imshow(image)
image2.set_title('RGB Image')

#Grayscale Image
image3.imshow(grayscale_image,cmap=plt.cm.get_cmap('gray'))
image3.set_title('Grayscale Image')
print(grayscale_image.shape)
plt.show()
```

```python
#HISTOGRAM EQUALIZATION
equalized_image = cv2.equalizeHist(image_gray)

plt.figure(figsize=(8, 4))

# Original Grayscale Image
plt.subplot(1, 2, 1)
plt.title("Original Grayscale Image")
plt.imshow(image_gray, cmap="gray")


# Equalized Grayscale Image
plt.subplot(1, 2, 2)
plt.title("Equalized Grayscale Image")
plt.imshow(equalized_image, cmap="gray")


plt.tight_layout()
plt.show()

# Histogram
plt.figure(figsize=(8, 4))

# Histogram for Original Image
plt.subplot(1, 2, 1)
plt.title("Histogram (Original Image)")
sns.histplot(image_gray.ravel(), bins=30, kde=True, color='skyblue', alpha=0.6)
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

# Histogram for Equalized Image
plt.subplot(1, 2, 2)
plt.title("Histogram (Equalized Image)")
sns.histplot(equalized_image.ravel(), bins=30, kde=True, color='skyblue', alpha=0.6)
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
#Image Resize
new_width=250
new_height=199
resize_img=cv2.resize(img,(new_width,new_height))

plt.figure(figsize=(10,7))
plt.subplot(121)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.subplot(122)
plt.imshow(cv2.cvtColor(resize_img,cv2.COLOR_BGR2RGB))
plt.title('Resized Image')
plt.show()
```

```python
#Adjust Contrast
a=1.7	#Contrast		
b=2	#Brightness	
adjusted_img=cv2.convertScaleAbs(img,alpha=a,beta=b)

plt.figure(figsize=(8,4))
plt.subplot(121)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.subplot(122)
plt.imshow(cv2.cvtColor(adjusted_img,cv2.COLOR_BGR2RGB))
plt.title('Contrast Adjusted Image')
plt.show()
```
